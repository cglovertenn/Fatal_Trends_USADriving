```python
stateinfodf = pd.read_csv('Resources/bad-drivers.csv')
stateinfodf
```

```python
state_groups = stateinfodf.groupby(['State'])

# Create a new variable that holds the sum of our groups
sum_it_up = state_groups.sum()
sum_it_up.head()
```

```python
state = "Louisiana"

just_one_state = sum_it_up.loc[state]

data_list = just_one_state.keys()

data_pie = just_one_state.plot(kind="pie", y=data_list, title=("Info on " + state))
data_pie.set_ylabel("")

plt.show()
plt.tight_layout()
plt.axis("equal")
```
