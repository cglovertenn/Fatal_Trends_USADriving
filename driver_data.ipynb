```python
import pandas as pd
import csv
```

```python
# Read CSV
driver_data = pd.read_csv("Resources/bad-drivers.csv")
```

```python
driver_data.head()
```

```python
sortby_billion = driver_data.sort_values("Number of drivers involved in fatal collisions per billion miles", ascending = False)
sortby_billion.head()
```

```python
sortby_speeding = driver_data.sort_values("Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding", ascending = False)
sortby_speeding.head()
```

```python
sortby_drinking = driver_data.sort_values("Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired", ascending = False)
sortby_drinking.head()
```

```python
sortby_distracted = driver_data.sort_values("Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted", ascending = True)
sortby_distracted.head()
```

```python
sortby_firsttime = driver_data.sort_values("Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents", ascending = False)
sortby_firsttime.head()
```

```python
sortby_premiums = driver_data.sort_values("Car Insurance Premiums ($)", ascending = False)
sortby_premiums.head()
```

```python
sortby_inslosses = driver_data.sort_values("Losses incurred by insurance companies for collisions per insured driver ($)", ascending = False)
sortby_inslosses.head()
```
